```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
pd.set_option("display.max_columns", None)
df=pd.read_csv(r"D:\internship\hr dataset\HR_comma_sep (1).csv")
df.head()
```

```python
df.isnull().sum()
```

```python
num_df=df.select_dtypes(exclude='object')
num_df
```

```python
cat_df=df.select_dtypes(include='object')
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinalencoder=OrdinalEncoder()
```

```python
encoded_data=ordinalencoder.fit_transform(cat_df)
```

```python
cat_encoded_df=pd.DataFrame(encoded_data,columns=['Department','salary'])
cat_encoded_df
```

```python
df_new=pd.merge(num_df,cat_encoded_df,left_index=True,right_index=True)
df_new
```

```python
X=df_new.drop(columns='left', axis=1)
X.head()
```

```python
y= df_new['left']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

regressor=LinearRegression()
```

```python
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
y_pred
```

```python
# score method used to calculate accuracy in classification models 
regressor.score(X_test,y_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
mean_absolute_error(y_test,y_pred)
```

```python
def evaluate_model(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2_square = r2_score(true, predicted)
    return mae, rmse, r2_square
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet

models={
    'LinearRegression':LinearRegression(),
    'Lasso':Lasso(),
    'Ridge':Ridge(),
    'Elasticnet':ElasticNet()
}
trained_model_list=[]
model_list=[]
r2_list=[]

for i in range(len(list(models))):
    model=list(models.values())[i]
    model.fit(X_train,y_train)

    #Make Predictions
    y_pred=model.predict(X_test)

    mae, rmse, r2_square=evaluate_model(y_test,y_pred)

    print(list(models.keys())[i])
    model_list.append(list(models.keys())[i])

    print('Model Training Performance')
    print("RMSE:",rmse)
    print("MAE:",mae)
    print("R2 score",r2_square*100)

    r2_list.append(r2_square)
    
    print('='*35)
    print('\n')
```
